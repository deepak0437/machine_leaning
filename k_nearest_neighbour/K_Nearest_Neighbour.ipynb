```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Networkads.csv')
df
```

```python
x = df.drop(['Purchased' , 'User ID'] , axis='columns')
y = df['Purchased']
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
x['Gender'] = encode.fit_transform(df['Gender'])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classi = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classi.fit(x_train,y_train)
```

```python
print(classi.predict(scaler.transform([[1,10,8000]])))
```

```python
y_pred = classi.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
